```python
import pandas as pd
import numpy as np

import requests
import json
from credenciales import cnbc_credentials, reuters_credentials 
```

#### URLs

```python
reuters_url = "https://reuters-business-and-financial-news.p.rapidapi.com/category-id/10/article-date/2023-10-05"
cnbc_url = "https://cnbc.p.rapidapi.com/news/v2/list-by-symbol"
```

```python

headers_reuters = {
	"X-RapidAPI-Key": reuters_credentials["X-RapidAPI-Key"],
	"X-RapidAPI-Host": reuters_credentials["X-RapidAPI-Host"]
}
```

```python
cnbc_headers = {
	"X-RapidAPI-Key": cnbc_credentials["X-RapidAPI-Key"],
	"X-RapidAPI-Host": cnbc_credentials["X-RapidAPI-Host"]
}
```

```python
response1 = requests.get(reuters_url , headers=headers_reuters)

print(response1.json())
```

```python
df_reuters = pd.DataFrame(json.loads(response1.text))
df_reuters
```

```python
df_reuters.shape
```

```python
querystring = {"symbol":"JPM","page":"1","pageSize":"30"}
response2 = requests.get(cnbc_url, headers=cnbc_headers, params=querystring)

print(response2.json())
```

```python
df = pd.DataFrame(json.loads(response2.text)['data']['symbolEntries']['results'])
df.head()
```
